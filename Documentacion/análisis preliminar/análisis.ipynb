## Variables Numéricas
### Análisis de la cantidad de valores válidos sobre el total, media y mediana, desviación estándar, valores ausentes y mal registrados o erróneos.


```python
import pandas as pd
import numpy as np

# Columnas numéricas objetivo del análisis
columnas_numericas = ['duracionAnios', 'cantidadCarrerasVirtuales', 'cantidadAreas']

# Convertir columnas a numérico sin imputar ni corregir valores
for col in columnas_numericas:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("📈 Análisis estadístico de columnas numéricas específicas:\n")

for col in columnas_numericas:
    print(f"📊 Columna: {col}")

    total = len(df)
    valores_validos = df[col].notnull().sum()
    ausentes = df[col].isnull().sum()
    media = df[col].mean()
    mediana = df[col].median()
    desviacion = df[col].std()

    negativos = (df[col] < 0).sum()
    infinitos = np.isinf(df[col]).sum()
    fuera_de_rango = ((df[col] > 1e6) | (df[col] < -1e6)).sum()

    print(f"  ✔️ Valores válidos: {valores_validos} / {total}")
    print(f"  ❗ Ausentes: {ausentes}")
    print(f"  📌 Media: {media:.2f}" if not np.isnan(media) else "  📌 Media: NaN")
    print(f"  📌 Mediana: {mediana:.2f}" if not np.isnan(mediana) else "  📌 Mediana: NaN")
    print(f"  📌 Desviación estándar: {desviacion:.2f}" if not np.isnan(desviacion) else "  📌 Desviación estándar: NaN")
    print(f"  ⚠️ Valores negativos: {negativos}")
    print(f"  ⚠️ Infinitos: {infinitos}")
    print(f"  ⚠️ Valores fuera de rango (+/- 1e6): {fuera_de_rango}")
    print("-" * 60)
```

> 🔎 **Nota sobre el análisis numérico**:
>
> Para realizar un análisis estadístico básico de las columnas numéricas del dataset (`duracionAnios`, `cantidadCarrerasVirtuales`, `cantidadAreas`), se ha convertido su contenido a tipo numérico utilizando `errors='coerce'`, lo que transforma los valores no convertibles en `NaN`.
>
> Esta conversión **no implica limpieza de datos ni corrección**, sino que permite identificar la cantidad de datos válidos y detectar valores ausentes, extremos o mal registrados para su posterior tratamiento en una etapa futura del proyecto.


## Variables Categóricas
### Análisis de la cantidad de registros para cada categoría, registros válidos sobre el total, ausentes y mal registrados.
    

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(14, 12))  # 2 filas x 2 columnas

# 1. Gráfico de barras: Cantidad de ausentes por variable
ausentes_por_variable = df.isnull().sum()
ausentes_por_variable = ausentes_por_variable[ausentes_por_variable > 0].sort_values(ascending=False)
if not ausentes_por_variable.empty:
    axs[0, 0].bar(ausentes_por_variable.index, ausentes_por_variable.values, color='tomato')
    axs[0, 0].set_title('Cantidad de valores ausentes por variable')
    axs[0, 0].set_ylabel('Cantidad de ausentes')
    axs[0, 0].tick_params(axis='x', rotation=45)
    for i, v in enumerate(ausentes_por_variable.values):
        axs[0, 0].text(i, v + 1, f"{v}", ha='center', va='bottom', fontsize=9)
else:
    axs[0, 0].text(0.5, 0.5, "🎉 No hay valores ausentes", ha='center', va='center', fontsize=14)
    axs[0, 0].axis('off')

# 2. Gráfico de Pareto para 'areaConocimiento'
columna_objetivo = "areaConocimiento"
if columna_objetivo in df.columns:
    conteo = df[columna_objetivo].value_counts()
    porcentaje_acumulado = conteo.cumsum() / conteo.sum()
    categorias = conteo.index[:15]
    valores = conteo.values[:15]
    porcentaje = porcentaje_acumulado.values[:15]

    ax1 = axs[0, 1]
    ax1.bar(np.arange(len(categorias)), valores, color='skyblue')
    ax1.set_ylabel('Frecuencia')
    ax1.set_xticks(np.arange(len(categorias)))
    ax1.set_xticklabels(categorias, rotation=45, ha='right')

    ax2 = ax1.twinx()
    ax2.plot(np.arange(len(categorias)), porcentaje, color='crimson', marker='o')
    ax2.set_ylabel('% Acumulado')
    ax2.set_ylim([0, 1.05])
    ax2.axhline(0.8, color='gray', linestyle='--', alpha=0.7)
    ax2.text(0, 0.82, "80%", color="gray")

    ax1.set_title(f"Gráfico de Pareto - {columna_objetivo}")
else:
    axs[0, 1].text(0.5, 0.5, f"⚠️ Columna '{columna_objetivo}' no encontrada", ha='center', va='center', fontsize=14)
    axs[0, 1].axis('off')

# 3. Gráfico circular: Proporción datos presentes vs ausentes
total_celdas = df.shape[0] * df.shape[1]
total_ausentes = df.isnull().sum().sum()
total_presentes = total_celdas - total_ausentes
sizes = [total_presentes, total_ausentes]
colors = ['#4CAF50', '#F44336']
explode = (0, 0.1)

axs[1, 0].pie(
    sizes,
    labels=['Presentes', 'Ausentes'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    explode=explode,
    shadow=True
)
axs[1, 0].set_title('Proporción de Datos Presentes vs Ausentes')
axs[1, 0].axis('equal')

# 4. Gráfico de barras: Cantidad de carreras por nivel
if 'nivel' in df.columns:
    conteo_nivel = df['nivel'].value_counts().sort_values(ascending=False)
    axs[1, 1].bar(conteo_nivel.index, conteo_nivel.values, color='mediumslateblue')
    axs[1, 1].set_title('Cantidad de carreras por nivel')
    axs[1, 1].set_ylabel('Cantidad de carreras')
    axs[1, 1].tick_params(axis='x', rotation=45)
    for i, v in enumerate(conteo_nivel.values):
        axs[1, 1].text(i, v + 1, f"{v}", ha='center', va='bottom', fontsize=9)
else:
    axs[1, 1].text(0.5, 0.5, "⚠️ La columna 'nivel' no existe", ha='center', va='center', fontsize=14)
    axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```
