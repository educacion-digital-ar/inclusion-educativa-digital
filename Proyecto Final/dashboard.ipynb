```python
# @title
# --- Preparar datos para los gr√°ficos ---

# Gr√°fico 1: P√∫blica vs Privada
oferta_tipo_real = datos_reales['sector'].value_counts().reset_index()
oferta_tipo_real.columns = ['Tipo de Instituci√≥n', 'Cantidad de Carreras Virtuales']

# Gr√°fico 2: Tipos de t√≠tulos para todos los sectores
# Preparamos los datos para los diferentes filtros (Todos, P√∫blica, Privada)
# Para sector p√∫blico
carreras_publicas = datos_reales[datos_reales['sector'] == 'P√∫blica']
tipos_titulo_publico = carreras_publicas['nivel'].value_counts().reset_index()
tipos_titulo_publico.columns = ['T√≠tulo', 'Cantidad']

# Para sector privado
carreras_privadas = datos_reales[datos_reales['sector'] == 'Privada']
tipos_titulo_privado = carreras_privadas['nivel'].value_counts().reset_index()
tipos_titulo_privado.columns = ['T√≠tulo', 'Cantidad']

# Para todos los sectores juntos
tipos_titulo_real = datos_reales['nivel'].value_counts().reset_index()
tipos_titulo_real.columns = ['T√≠tulo', 'Cantidad']

# Gr√°fico 3: Distribuci√≥n regional de carreras p√∫blicas virtuales
# Usamos el DataFrame de carreras p√∫blicas filtrado y la columna 'provincia'
regiones_real = carreras_publicas['provincia'].value_counts().reset_index()
regiones_real.columns = ['Regi√≥n', 'Carreras P√∫blicas Virtuales']
```

```python
# @title
import plotly.graph_objects as go
import pandas as pd
from plotly.subplots import make_subplots
from IPython.display import display, Markdown, HTML
import ipywidgets as widgets
import plotly.io as pio
display(HTML("""
<style>
.jp-OutputArea-output, .output_wrapper, .output, .plotly-graph-div {
    display: flex !important;
    justify-content: center !important;
    align-items: center !important;
    margin-left: auto !important;
    margin-right: auto !important;
}
</style>
"""))

# Definir las paletas de colores
paletas = {
    'Violeta': {
        'bar': ['#6C5B7B', '#3F357D'],
        'pie': ['#6C5B7B', '#3F357D', '#5B5F97', '#8D6EC4', '#9778D3', '#7B68EE', '#A470B8', '#7851A9',
                '#614E6E', '#483D8B', '#7D6B91', '#9966CC', '#9370DB', '#B19CD9', '#8A2BE2', '#4B0082'],
        'region': ['#6C5B7B', '#3F357D', '#5B5F97', '#8D6EC4', '#4056A1', '#593A5A', '#8B53C8', '#483D8B', '#9370DB', '#663399',
                   '#301934', '#7854A6', '#5D3FD3', '#A679D2', '#BDB2FF', '#9A85FF', '#D6CAFF', '#6247AA']
    },
    'Rojos': {
        'bar': ['#B71C1C', '#D32F2F'],
        'pie': ['#B71C1C', '#D32F2F', '#F44336', '#E57373', '#FF8A80', '#FF5252', '#C62828', '#FF1744',
                '#8F0000', '#E53935', '#FFCCCB', '#FF6B6B', '#EF5350', '#FFA4A2', '#D00000', '#DC143C'],
        'region': ['#B71C1C', '#D32F2F', '#F44336', '#E57373', '#EF5350', '#C62828', '#FF9E80', '#FF6E40', '#BF360C', '#DD2C00',
                   '#B91C1C', '#B22222', '#E25822', '#CF1020', '#A52A2A', '#990000', '#CD5C5C', '#E34234']
    },
    'Celestes/Verdes': {
        'bar': ['#0097A7', '#00796B'],
        'pie': ['#0097A7', '#00796B', '#00ACC1', '#26A69A', '#4DB6AC', '#80DEEA', '#00BFA5', '#1DE9B6',
                '#006064', '#004D40', '#00838F', '#00695C', '#00BCD4', '#009688', '#80CBC4', '#84FFFF'],
        'region': ['#0097A7', '#00796B', '#00ACC1', '#26A69A', '#00BCD4', '#009688', '#4DD0E1', '#00E5FF', '#00B8D4', '#2E7D32',
                   '#00897B', '#4DB6AC', '#00766C', '#48A999', '#00A693', '#007F73', '#00BFA5', '#1DE9B6']
    },
    'Varios': {
        'bar': ['#FF5722', '#03A9F4'],
        'pie': ['#FF5722', '#03A9F4', '#FFC107', '#8BC34A', '#673AB7', '#CDDC39', '#009688', '#3F51B5',
                '#FF9800', '#2196F3', '#FFEB3B', '#4CAF50', '#9C27B0', '#8BC34A', '#00BCD4', '#F44336'],
        'region': ['#FF5722', '#03A9F4', '#FFC107', '#8BC34A', '#9C27B0', '#E91E63', '#4CAF50', '#2196F3', '#FF9800', '#795548',
                   '#607D8B', '#3F51B5', '#FFEB3B', '#CDDC39', '#673AB7', '#009688', '#F44336', '#00BCD4']
    }
}


def actualizar_grafico(vista, paleta_seleccionada, modo, filtro_sector):
    # Selecciona la paleta de colores actual
    current_paleta = paletas[paleta_seleccionada]
    current_bar_colors = current_paleta['bar']
    current_pie_colors = current_paleta['pie']
    current_region_colors = current_paleta['region']

    # Selecciona los datos para el segundo gr√°fico seg√∫n el filtro
    if filtro_sector == 'P√∫blica':
        tipos_titulo = tipos_titulo_publico
    elif filtro_sector == 'Privada':
        tipos_titulo = tipos_titulo_privado
    else:
        tipos_titulo = tipos_titulo_real

    # Define colores de fondo y texto seg√∫n el modo
    if modo == 'oscuro':
        bg_color = '#303030'
        text_color = '#E0E0E0'
        grid_color = '#505050'
    else:
        bg_color = '#FFFFFF'
        text_color = '#212121'
        grid_color = '#E0E0E0'

    # Ajustar altura seg√∫n el tipo de vista
    height = 900 if vista == "Torta" else 800

    # Crear subplots con m√°s espacio entre ellos
    fig = make_subplots(
        rows=2, cols=2,
        specs=[
            [{"type": "bar" if vista == "Barras" else "pie"}, {"type": "bar"}],
            [{"colspan": 2, "type": "bar" if vista == "Barras" else "pie"}, None]
        ],
        subplot_titles=(
            "üéì P√∫blica vs Privada",
            "üèÖ Tipos de t√≠tulos p√∫blicos virtuales",
            "üó∫Ô∏è Distribuci√≥n regional de carreras p√∫blicas virtuales",
            ""
        ),
        vertical_spacing=0.30,  # Aumentado para m√°s espacio entre filas
        horizontal_spacing=0.15  # Aumentado para m√°s espacio entre columnas
    )

    # Gr√°fico 1: Instituci√≥n
    if vista == "Barras":
        fig.add_trace(
            go.Bar(
                x=oferta_tipo_real['Tipo de Instituci√≥n'],
                y=oferta_tipo_real['Cantidad de Carreras Virtuales'],
                text=oferta_tipo_real['Cantidad de Carreras Virtuales'],
                textposition='outside',
                marker_color=[current_bar_colors[0] if t == 'P√∫blica' else current_bar_colors[1] for t in oferta_tipo_real['Tipo de Instituci√≥n']],
                showlegend=False,
                marker=dict(line=dict(width=0.5, color=bg_color))  # Bordes m√°s finos
            ),
            row=1, col=1
        )
    else:
        fig.add_trace(
            go.Pie(
                labels=oferta_tipo_real['Tipo de Instituci√≥n'],
                values=oferta_tipo_real['Cantidad de Carreras Virtuales'],
                hole=0.4,
                textinfo='label+percent',
                marker=dict(
                    colors=[current_bar_colors[0] if t == 'P√∫blica' else current_bar_colors[1] for t in oferta_tipo_real['Tipo de Instituci√≥n']],
                    line=dict(color=bg_color, width=1)
                ),
                showlegend=False,
                textfont=dict(size=14)
            ),
            row=1, col=1
        )

    # Gr√°fico 2: Tipos de t√≠tulos
    fig.add_trace(
        go.Bar(
            x=tipos_titulo['T√≠tulo'],
            y=tipos_titulo['Cantidad'],
            text=tipos_titulo['Cantidad'],
            textposition='outside',
            marker_color=current_pie_colors[:len(tipos_titulo)],
            showlegend=False,
            marker=dict(line=dict(width=0.5, color=bg_color))  # Bordes m√°s finos
        ),
        row=1, col=2
    )

    # Gr√°fico 3: Regiones
    if vista == "Barras":
        fig.add_trace(
            go.Bar(
                x=regiones_real['Regi√≥n'],
                y=regiones_real['Carreras P√∫blicas Virtuales'],
                text=regiones_real['Carreras P√∫blicas Virtuales'],
                textposition='outside',
                marker_color=current_region_colors[:len(regiones_real)],
                showlegend=False,
                marker=dict(line=dict(width=0.5, color=bg_color))  # Bordes m√°s finos
            ),
            row=2, col=1
        )
    else:
        fig.add_trace(
            go.Pie(
                labels=regiones_real['Regi√≥n'],
                values=regiones_real['Carreras P√∫blicas Virtuales'],
                hole=0.4,
                textinfo='label+percent',
                marker=dict(
                    colors=current_region_colors[:len(regiones_real)],
                    line=dict(color=bg_color, width=1)
                ),
                showlegend=False,
                textfont=dict(size=14)
            ),
            row=2, col=1
        )

    # Layout mejorado
    fig.update_layout(
        autosize=True,
        height=height,
        margin=dict(l=60, r=60, t=120, b=100),  # M√°rgenes aumentados
        title={
            'text': "üìä Dashboard de la oferta de carreras virtuales en Argentina",
            'x': 0.5,
            'y': 0.98,
            'xanchor': 'center',
            'font': dict(size=18, color=text_color, family="Arial, sans-serif")
        },
        font=dict(family="Arial, sans-serif", size=14, color=text_color),
        plot_bgcolor=bg_color,
        paper_bgcolor=bg_color,
        hovermode='closest',
        uniformtext=dict(minsize=12, mode='hide')  # Texto uniforme
    )

    # Subt√≠tulos m√°s grandes y con mejor espaciado
    for i, ann in enumerate(fig['layout']['annotations']):
        ann['font']['color'] = text_color
        ann['font']['size'] = 16
        ann['font']['family'] = "Arial, sans-serif"
        # Ajustar posici√≥n vertical de los t√≠tulos
        if i < 2:  # Subt√≠tulos de la primera fila
            ann['y'] = ann['y'] + 0.03
        else:  # Subt√≠tulo de la segunda fila
            ann['y'] = ann['y'] + 0.02

    # Ajustar rangos de ejes para gr√°ficos de barras
    if vista == "Barras":
        max_oferta = oferta_tipo_real['Cantidad de Carreras Virtuales'].max() if not oferta_tipo_real.empty else 0
        max_titulo = tipos_titulo['Cantidad'].max() if not tipos_titulo.empty else 0
        max_region = regiones_real['Carreras P√∫blicas Virtuales'].max() if not regiones_real.empty else 0

        fig.update_yaxes(range=[0, max_oferta * 1.3], row=1, col=1, gridcolor=grid_color, showgrid=True, title_text="Cantidad", title_standoff=15)
        fig.update_yaxes(range=[0, max_titulo * 1.3], row=1, col=2, gridcolor=grid_color, showgrid=True, title_text="Cantidad", title_standoff=15)
        fig.update_yaxes(range=[0, max_region * 1.3], row=2, col=1, gridcolor=grid_color, showgrid=True, title_text="Carreras", title_standoff=15)

        fig.update_xaxes(title_text="", row=1, col=1, gridcolor=grid_color, showgrid=False)
        fig.update_xaxes(title_text="", row=1, col=2, gridcolor=grid_color, showgrid=False)
        fig.update_xaxes(title_text="", row=2, col=1, gridcolor=grid_color, showgrid=False)
    else:
        # Para gr√°ficos de torta, ocultar ejes
        fig.update_xaxes(showgrid=False, zeroline=False, showticklabels=False)
        fig.update_yaxes(showgrid=False, zeroline=False, showticklabels=False)

        # Ajustar tama√±o de las tortas para que sean m√°s grandes
        fig.update_layout(
            polar=dict(
                domain=dict(x=[0, 1], y=[0, 1])
            )
        )

    fig.show()

    display(HTML("""
    <style>
    .output_wrapper, .output, .plotly-graph-div {
        width: 100% !important;
        max-width: 100% !important;
        box-sizing: border-box;
        overflow-x: hidden !important;
    }
    .pie-chart {
        transform: scale(1.2);
        transform-origin: center center;
    }
    </style>
    """))


# --- Widgets Interactivos ---
display(Markdown("<h2 style='text-align:center; margin-top: 30px; margin-bottom: 20px; font-family: Arial, sans-serif;'>Controles del Dashboard</h2>"))

vista_tipo = widgets.ToggleButtons(
    options=['Barras', 'Torta'],
    value='Barras',
    description='Vista:',
    style={'description_width': 'initial', 'button_width': '100px'},
    button_style='info'
)

paleta_selector = widgets.ToggleButtons(
    options=['Violeta', 'Rojos', 'Celestes/Verdes', 'Varios'],
    value='Violeta',
    description='Colores:',
    style={'description_width': 'initial', 'button_width': '120px'},
    button_style='info'
)

modo_selector = widgets.ToggleButtons(
    options=[('üåû', 'claro'), ('üåô', 'oscuro')],
    value='claro',
    description='Modo:',
    style={'description_width': 'initial', 'button_width': '80px'},
    button_style='info'
)

filtro_sector = widgets.Dropdown(
    options=[('Ambos', 'Todos'), ('P√∫blica', 'P√∫blica'), ('Privada', 'Privada')],
    value='Todos',
    description='Sector:',
    style={'description_width': 'initial'},
)

# Organizar widgets con mejor espaciado
controls_container = widgets.VBox([
    widgets.HBox([vista_tipo, paleta_selector], layout=widgets.Layout(justify_content='space-around', padding='10px')),
    widgets.HBox([modo_selector, filtro_sector], layout=widgets.Layout(justify_content='space-around', padding='10px'))
], layout=widgets.Layout(margin='20px 0px', padding='10px', border='1px solid #eee', border_radius='8px'))

display(controls_container)

out = widgets.interactive_output(
    actualizar_grafico,
    {
        'vista': vista_tipo,
        'paleta_seleccionada': paleta_selector,
        'modo': modo_selector,
        'filtro_sector': filtro_sector
    }
)

display(out)
```

## 3. An√°lisis y hallazgos clave

### 1. Predominio del sector privado
La gran mayor√≠a de las carreras virtuales ofrecidas en Argentina (88.1%) corresponden a universidades privadas. Esto evidencia que el sector privado lidera ampliamente la oferta de educaci√≥n superior a distancia, mientras que el sector p√∫blico tiene una participaci√≥n mucho menor en este √°mbito.

### 2. Diversidad en los tipos de t√≠tulos
Los t√≠tulos m√°s ofrecidos en modalidad virtual son los de pregrado y grado, especialmente tecnicaturas y licenciaturas. Sin embargo, tambi√©n hay una oferta relevante de posgrados, lo que muestra que la virtualidad est√° presente en todos los niveles de formaci√≥n.

### 3. Concentraci√≥n regional
Buenos Aires lidera con una diferencia muy amplia en la cantidad de carreras p√∫blicas virtuales (por ejemplo, 48), seguida por C√≥rdoba (17) y Santiago del Estero (7). Esta concentraci√≥n puede estar vinculada a la presencia de universidades nacionales y a una mejor infraestructura tecnol√≥gica en estas provincias.

### 4. Brecha regional
Existen provincias con una oferta virtual p√∫blica muy baja o nula, lo que evidencia una brecha territorial en el acceso a la educaci√≥n superior a distancia. Esto puede deberse a factores como menor densidad poblacional, menor infraestructura tecnol√≥gica o menor presencia de universidades p√∫blicas.

Estos hallazgos revelan una brecha estructural en el acceso a la educaci√≥n superior, donde el lugar de residencia y la capacidad de pago siguen siendo factores determinantes.


## 4. Conclusiones y propuestas

Como conclusion el an√°lisis evidenci√≥ una profunda desigualdad en el acceso a la educaci√≥n superior p√∫blica en modalidad virtual en Argentina. La oferta de carreras virtuales es limitada y se concentra en unas pocas jurisdicciones, lo que impacta negativamente en poblaciones que enfrentan barreras geogr√°ficas, econ√≥micas o familiares para acceder a la modalidad presencial. Asimismo, se observ√≥ una marcada preponderancia de instituciones privadas en la oferta acad√©mica virtual, lo cual puede agravar la exclusi√≥n educativa en los sectores m√°s vulnerables. Esta situaci√≥n requiere una intervenci√≥n urgente desde el √°mbito de las pol√≠ticas p√∫blicas, con decisiones basadas en datos y centradas en garantizar el derecho a una educaci√≥n superior inclusiva, equitativa y federal
