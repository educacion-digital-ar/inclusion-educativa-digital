```python
import pandas as pd

sheet_id = '1Qa5X6_s5zTJIFG6y9yaMngnWy01OKwOY'
sheet_name = 'Hoja1'

url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'

df = pd.read_csv(url)

print("Primeros 10 registros:")
print(df.head(10))


```

   En las columnas numéricas analizar la cantidad de valores válidos sobre el total, media y mediana, desviación estándar, valores ausentes y mal registrados o erróneos.


```python
import numpy as np


columnas_numericas = ['duracionAnios', 'cantidadCarrerasVirtuales', 'cantidadAreas']

# Convertir a numérico, convirtiendo errores a NaN
for col in columnas_numericas:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("📈 Análisis estadístico de columnas numéricas específicas:\n")

for col in columnas_numericas:
    print(f"📊 Columna: {col}")

    total = len(df)
    valores_validos = df[col].notnull().sum()
    media = df[col].mean()
    mediana = df[col].median()
    desviacion = df[col].std()
    ausentes = df[col].isnull().sum()

    negativos = (df[col] < 0).sum()
    infinitos = np.isinf(df[col]).sum()
    fuera_de_rango = ((df[col] > 1e6) | (df[col] < -1e6)).sum()

    print(f"  ✔️ Valores válidos: {valores_validos} / {total}")
    print(f"  📌 Media: {media:.2f}" if not np.isnan(media) else "  📌 Media: NaN")
    print(f"  📌 Mediana: {mediana:.2f}" if not np.isnan(mediana) else "  📌 Mediana: NaN")
    print(f"  📌 Desviación estándar: {desviacion:.2f}" if not np.isnan(desviacion) else "  📌 Desviación estándar: NaN")
    print(f"  ❗ Ausentes: {ausentes}")
    print(f"  ⚠️ Negativos: {negativos}")
    print(f"  ⚠️ Infinitos: {infinitos}")
    print(f"  ⚠️ Valores fuera de rango (+/- 1e6): {fuera_de_rango}")
    print("-" * 60)
```

    En las variables categóricas analizar la cantidad de registros para cada categoría, registros válidos sobre el total, ausentes y mal registrados.
    

```python
import pandas as pd
import os
from IPython.display import display
import unicodedata
import matplotlib.pyplot as plt

resultados = []

def normalizar_texto(texto):
    if pd.isnull(texto):
        return texto
    texto = str(texto).strip().lower()
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    return texto

for c in df.columns:
    if df[c].dtype == 'object' or df[c].dtype.name == 'category':
        df[c] = df[c].apply(normalizar_texto)
        total = len(df)
        validas = df[c].notnull().sum()
        ausentes = df[c].isnull().sum()
        error_registro = df[c].apply(lambda x: not isinstance(x, str) if pd.notnull(x) else False).sum()
        cantidad_valores = df[c].value_counts(dropna=False)

        resultados.append({ "Variable" : c,
                           "Total registrados": total,
                            "Válidos" : validas,
                            "Ausentes" : ausentes,
                            "Mal registrados" : error_registro,
                            "Distribucion de valores" : cantidad_valores})


resultados_df = pd.DataFrame(resultados)

display(resultados_df)

for col in df.columns:
    if df[col].dtype == 'object' or df[col].dtype.name == 'category':
        plt.figure(figsize=(10, 5))
        df[col].value_counts().head(10).plot(kind='bar', color='salmon')
        plt.title(f'Distribución de valores para "{col}"')
        plt.ylabel("Cantidad")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

print("\n--- Carreras por porcentaje de modalidad virtual ---\n")
agrupado = df.groupby("porcentajeOfertaVirtual")["nombreCarrera"].unique()

for porcentaje, carreras in agrupado.items():
    print(f"🔹 {porcentaje}:")
    print(", ".join(carreras[:10]))
    print()
```
